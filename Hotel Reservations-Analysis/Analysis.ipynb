```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/cleaned_hotel_booking.csv")

sns.set(style="whitegrid", context="talk", palette="deep")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Booking Status â€“ Confirmed vs Cancelled
plt.figure(figsize=(6,6))
df['booking_status'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Booking Status Breakdown")
plt.ylabel("")
plt.show()
```

```python
# Room Price Distribution
plt.figure(figsize=(4,4))
sns.histplot(df["avg_price_per_room"], kde=True, edgecolor="black")
plt.title("Average Price Per Room â€“ Distribution", fontsize=16, fontweight="bold")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Adults Count Distribution
plt.figure(figsize=(5,5))
sns.histplot(df['no_of_adults'], kde=True)
plt.title("Distribution of Number of Adults")
plt.show()
```

```python
#Meal Plan Usage
plt.figure(figsize=(6,6))
df['type_of_meal_plan'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Meal Plan Distribution")
plt.ylabel("")
plt.show()
```

```python
#Room Types Reserved
plt.figure(figsize=(5,5))
sns.countplot(y=df['room_type_reserved'], order=df['room_type_reserved'].value_counts().index)
plt.title("Room Types Reserved")
plt.show()
```

```python
#Lead Time Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['lead_time'], kde=True)
plt.title("Lead Time Distribution")
plt.show()
```

```python
# Arrival Month Trends
plt.figure(figsize=(8,5))
df['arrival_month'].value_counts().sort_index().plot(kind='bar')
plt.title("Bookings by Month")
plt.xlabel("Month")
plt.ylabel("Bookings Count")
plt.show()
```

```python
# Arrival Date Pattern
plt.figure(figsize=(8,4))
df['arrival_date'].value_counts().sort_index().plot(kind='line')
plt.title("Bookings by Check-in Date")
plt.xlabel("Day of Month")
plt.ylabel("Bookings Count")
plt.show()
```

```python
# Market Segment Types
plt.figure(figsize=(5,5))
sns.countplot(data=df, y='market_segment_type', order=df['market_segment_type'].value_counts().index)
plt.title("Market Segment Contribution")
plt.show()
```

```python
# Special Requests
plt.figure(figsize=(6,6))
sns.countplot(x=df['no_of_special_requests'])
plt.title("Count of Special Requests")
plt.show()
```
